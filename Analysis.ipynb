```python
# Load the CSV data

import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("data/users.csv")
df.head()
```

```python
# Clean & prepare the data

numeric_cols = [
    "Age", "Income",
    "Utilities", "Entertainment",
    "SchoolFees", "Shopping", "Healthcare"
]

for col in numeric_cols:
    df[col] = pd.to_numeric(df[col], errors="coerce")
```

```python
# Show ages with the highest income

age_income = df.groupby("Age")["Income"].mean().sort_values(ascending=False)

plt.figure()
age_income.head(10).plot(kind="bar")
plt.title("Ages with Highest Average Income")
plt.xlabel("Age")
plt.ylabel("Average Income")
plt.tight_layout()
plt.show()

# Export Chat

plt.figure()
age_income.head(10).plot(kind="bar")
plt.title("Ages with Highest Average Income")
plt.xlabel("Age")
plt.ylabel("Average Income")
plt.tight_layout()
plt.savefig("age_highest_income.png")
plt.close()

```

```python
# Gender distribution across spending categories

spending_cols = [
    "Utilities", "Entertainment",
    "SchoolFees", "Shopping", "Healthcare"
]

gender_spending = df.groupby("Gender")[spending_cols].sum()

plt.figure()
gender_spending.plot(kind="bar")
plt.title("Gender Distribution Across Spending Categories")
plt.xlabel("Gender")
plt.ylabel("Total Spending")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()

# Export chart

plt.figure()
gender_spending.plot(kind="bar")
plt.title("Gender Distribution Across Spending Categories")
plt.xlabel("Gender")
plt.ylabel("Total Spending")
plt.xticks(rotation=0)
plt.tight_layout()
plt.savefig("gender_spending_distribution.png")
plt.close()
```
